### Cleaning Data

Proses cleaning data juga dilakukan satu per satu pada dataframe setiap stasiun agar proses cleaning data dilakukan sesuai dengan karakteristik lokal data dari daerah dimana stasiun berada.

**Cleaning data Stasiun Aotizhongxin (AZN)**

Berdasarkan hasil asses data, dataframe ini memilki data kosong pada parameter utama data kualitas udara sehingga diperlukan proses imputasi.

```python
azn_df.isna().sum()
```

Terlihat bahwa perlu dilakukan imputasi pada kolom PM2.5, PM10, SO2, NO2, CO, O3, TEMP, PRES, DEWP, RAIN, WSPM, dan wd. Imputasi akan dilakukan dengan metode interpolasi untuk mengisi data yang kosong berdasarkan data-data lain yang terisi sehingga nilainya tetap representatif dan tidak menimbulkan anomali. Khusus untuk parameter wd, imputasi dilakukan dengan menginputkan nilai mayoritas karena bukan merupakan nilai numerik, sehingga tidak bisa dilakukan interpolasi

```python
miss_val = ['PM2.5','PM10','SO2','NO2','CO','O3','TEMP','PRES','DEWP','RAIN','WSPM']

for i in miss_val:
  azn_df[i] = azn_df[i].interpolate(method='linear')

azn_df.isna().sum()
```

```python
azn_df.wd.fillna(azn_df.wd.mode()[0], inplace=True)
azn_df.isna().sum()
```

```python
azn_df.describe()
```

**Insight:**
- Imputasi dengan metode interpolasi berhasil dilakukan dan tidak menimbulkan anomali pada parameter statistiknya
- Kolom wd diimputasi dengan modus karena tidak dapat diinterpolasi

## Analisis Lanjutan (Opsional)

Selanjutnya dilakukan analisis lanjutan geospasial untuk bisa lebih memeberi pemahaman akan hasil insight yang telah dipaparkan dari hasil ekplorasi data

Pertama, perlu dicari koordinat lokasi dari masing-masing stasiun menggunakan `geopy`

```python
!pip install geopy folium
```

```python
from geopy.geocoders import Nominatim
import time

# Membaca data
df = AQ_combined_df
# Inisialisasi geolocator
geolocator = Nominatim(user_agent="AirQualityAnalysisApp")

# Fungsi untuk mendapatkan koordinat berdasarkan nama stasiun
def get_coordinates(station_name):
    try:
        location = geolocator.geocode(station_name, timeout=10)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error: {e}")
        return None, None

# Mengambil koordinat untuk setiap stasiun
coordinates = []
for station in df['station'].unique():
    latitude, longitude = get_coordinates(station)
    coordinates.append((station, latitude, longitude))
    # Delay antara request untuk menghindari rate limit
    time.sleep(1)

# Mengubah hasil ke dalam DataFrame
coordinates_df = pd.DataFrame(coordinates, columns=['station', 'latitude', 'longitude'])

# Menampilkan hasil
print(coordinates_df)

# Menyimpan hasil ke CSV
coordinates_df.to_csv('stasiun_koordinat.csv', index=False)
```
